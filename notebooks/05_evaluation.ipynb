```python
import pandas as pd
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error
from src import prophet_visualization
```

## **1. Visualizing Actual vs Predicted Sales**

```python
df_eval = pd.read_csv('../data/raw/sales_train_evaluation.csv')
df_forecast = pd.read_parquet('../models/forecast/forecast_results.parquet')
cal = pd.read_csv('../data/raw/calendar.csv')
```

```python
df_compare = prophet_visualization.plot_actual_vs_predicted(df_forecast, df_eval, cal)
```

## **2. Metrics**

```python
mse = mean_squared_error(df_compare['actual'], df_compare['yhat'])
rmse = root_mean_squared_error(df_compare['actual'], df_compare['yhat'])
mae = mean_absolute_error(df_compare['actual'], df_compare['yhat'])

print('MSE:', mse)
print('RMSE:', rmse)
print('MAE:', mae)
```
